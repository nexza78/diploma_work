```python
import numpy as np
```

```python
sum = np.zeros((3, 3))
for i in range(10):
    tmp = np.load("cf_" + str(i) + ".npy")
    t1 = tmp/np.transpose(tmp.astype(float).sum(axis=1)).reshape(3, 1)*100
    sum += t1
```

```python
sum/10
```

```python
import seaborn as sns
sns.set(font_scale=1.5)
sns.heatmap(sum/1000, annot=True, fmt='.2%', cmap='Blues')
```
